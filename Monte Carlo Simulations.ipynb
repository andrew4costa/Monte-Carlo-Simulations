# Monte Carlo Simulations

A monte carlo simulation is a way to simulate random events in the manner of a virtual casino, except, it eliminates any biases that one might have in the physical world, like the speed of a ball on the roulette table and other external forces

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# 1. Coin Flip

```python
# define a function for a coin flip
def coin_flip():
    return random.randint(0,1)
```

```python
# list to store probability values
list1 = []

# define monte carlo function
def monte_carlo(n):
    results = 0
    for i in range(n):
        flip_result = coin_flip()
        results = results + flip_result
        
        # calculate frequentist probability
        probs = results/(i+1)
        
        # append to list
        list1.append(probs)
        
        # Plot result
        plt.axhline(y = 0.5, color = 'r', linestyle = '-') # marks the expected value on graph
        plt.xlabel("Iterations")
        plt.ylabel("Probability")
        plt.plot(list1)
        
    return results/n
```

```python
solution = monte_carlo(100)
print(f"Final Value after 100 simulations: {solution}")
```

Running the monte carlo simulation for 100 iterations, we can see a sharp drop in the probability, with  the probability of heads being less than 0.3 after 30 iterations. From this point onwards, the probability oscillates wildly, with the final probability after 100 iterations being 0.61.

```python
solution_1 = monte_carlo(1000)
print(f"Final value after 1000 iterations: {solution_1}")
```

The pattern follows a similar one to the simulation before, with the probability oscillating wildly until it begins to settle and converge close to 0.51 and the true value of 0.5

# 2. Estimating Pi using a Circle and a Square

Approach: 
- In order to estimate Pi, we need to determine the area of the square and circle
- To find these areas, dots will be fired randomly on the surface
- The dots that fall within the circle and outide the circle will then be counted
- The counts will then provide estimations of the areas of the circle and square

```python
import turtle
import math
```

```python
# Visualise random points
pen = turtle.Turtle()
pen.hideturtle()
pen.speed(0)

# draw square
pen.up()
pen.setposition(-100,-100)
pen.down()
pen.fd(200)
pen.left(90)
pen.fd(200)

pen.left(90)
pen.fd(200)
pen.left(90)
pen.fd(200)
pen.left(90)

# draw circle
pen.up()
pen.setposition(0,-100)
pen.down()
pen.circle(100)

# count points in/out of circle
in_circle = 0
out_circle = 0

# pi values
pi = []

for i in range(5):
    for j in range(1000):
        
        # generate random numbers
        x = random.randrange(-100,100)
        y = random.randrange(-100,100)
        
        # check if number lies outside circle
        if (x**2+y**2 > 100**2):
            pen.color("black")
            pen.up()
            pen.goto(x,y)
            pen.down()
            pen.dot()
            out_circle = out_circle + 1
        
        # if number lie inside circle
        else:
            pen.color("red")
            pen.up()
            pen.goto(x,y)
            pen.down()
            pen.dot()
            in_circle = in_circle + 1
        
        # calculate pi
        pi_ = 4.0 * in_circle / (in_circle + out_circle)
        
        # append values of PI to list
        pi.append(pi_)
        
        # calculate errors
        av_pi_errors = [abs(math.pi - pi) for pi in pi]
    
    print(pi[-1])
```

```python
plt.axhline(y = math.pi, color = 'g', linestyle= '-')
plt.plot(av_pi_errors)
plt.plot(pi)
plt.xlabel('Iterations')
plt.ylabel('Value of Pi')
plt.show()
```

```python
def dice_roll():
    return random.randint(1,6)
```

```python
dice_roll()
```

```python
list2 = []

def dice_monte(n):
    results = 0
    for i in range(n):
        roll_result = dice_roll()
        results = results + roll_result
        probs = results/(i+1)
        
        list2.append(probs)
        
        # expected value is 3.5
        plt.axhline(y = 3.5, color = 'g', linestyle = '-')
        plt.xlabel('Iterations')
        plt.ylabel('Probability')
        plt.title('Dice Roll')
        plt.plot(list2)
        
    return results/n
```

```python
dice_monte(1000)
```
